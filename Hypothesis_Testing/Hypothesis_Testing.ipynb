## T Test

A t-test is a type of inferential statistic which is used to determine if there is a significant difference between the means of two groups which may be related in certain features

T-test has 2 types : 1. one sampled t-test 2. two-sampled t-test.

##  One-sample T-test with Python

```python
import scipy.stats as stats
import numpy as np

ages=[10,20,35,50,28,40,55,18,16,55,30,25,43,18,30,28,14,24,16,17,32,35,26,27,65,18,43,23,21,20,19,70]
print(f"Length of ages data:{len(ages)}")
ages_mean=np.mean(ages)
print(ages_mean)
```

```python
## Lets take sample
sample_size=10
age_sample=np.random.choice(ages,sample_size)
age_sample
```

```python
from scipy.stats import ttest_1samp #one sample t testing
ttest,p_value=ttest_1samp(age_sample,30) #assumeing the 30 is the mean value of the sample
print(p_value)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("we are accepting null hypothesis")
```

## Some More Examples
Consider the age of students in a college and in Class A

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import math
np.random.seed(6)
school_ages=stats.poisson.rvs(loc=18,mu=35,size=1500) #total students=1500,age starts from 18 and mean=35
classA_ages=stats.poisson.rvs(loc=18,mu=30,size=60) #consider class A

classA_ages.mean()
```

```python
_,p_value=stats.ttest_1samp(a=classA_ages,popmean=school_ages.mean())#checking the class A mean against the school mean
p_value
```

```python
school_ages.mean()
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")#there is no age diffence b/w school and class A
else:
    print("we are accepting null hypothesis")
```

## Two-sample T-test With Python

```python
classB_ages=stats.poisson.rvs(loc=18,mu=33,size=60) #consider class A
classB_ages.mean()
```

```python
_,p_value=stats.ttest_ind(a=classA_ages,b=classB_ages,equal_var=False)
print(p_value)
```

```python
if p_value<0.05:  # alpha value is 0.05 or 5%
    print("Reject null hypothesis")
else:
    print("Do not Reject null hypothesis")
```

## Paired T-test With Python

```python
weight1=[25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
weight2=weight1+stats.norm.rvs(scale=5,loc=-1.25,size=15) #adding valuve to weight1 for 2nd sample
print(weight1)
print(weight2)
```

```python
#now lets compare weights
weight_df=pd.DataFrame({"Weight_01":np.array(weight1),
                       "Weight_02":np.array(weight2),
                       "Weight_changes":np.array(weight2)-np.array(weight1)})
weight_df

```

```python
_,p_value=stats.ttest_rel(a=weight1,b=weight2)
print(p_value)
```

```python
if p_value<0.05:
    print("Reject null hypothisis")
else:
    print("Do not reject null hypothisis")
```

## Chi-Square Test

```python
from scipy.stats import chi2_contingency, chi2
import numpy as np

# 2010 census proportions
proportions_2010 = [0.20, 0.30, 0.50]   # <50, 50-70, >70-->20%,30%,50%

# 2020 observed counts
observed = np.array([140, 160, 200])

n = 500   # total sample size 2020

# Expected counts from 2010 proportions × n
expected = np.array([p * n for p in proportions_2010])
```

```python
#CHI-SQUARE TEST
chi2_stat = np.sum((observed - expected) ** 2 / expected)
dof       = len(observed) - 1      # k - 1 = 3 - 1 = 2 #degree of freedom
alpha     = 0.05 # rejected area
critical  = chi2.ppf(1 - alpha, dof) 
p_value   = 1 - chi2.cdf(chi2_stat, dof)

critical
```

```python
print("=" * 55)
print("   CHI-SQUARE GOODNESS OF FIT TEST — CENSUS DATA")

print(f"\n{'Age Group':<12} {'Observed':>10} {'Expected':>10} {'(O-E)²/E':>10}")
print("-" * 45)
labels = ["< 50", "50 - 70", "> 70"]
for label, o, e in zip(labels, observed, expected):
    contrib = (o - e) ** 2 / e
    print(f"{label:<12} {o:>10} {e:>10.1f} {contrib:>10.2f}")
print("-" * 45)
print(f"{'Total':<12} {sum(observed):>10} {sum(expected):>10.1f} {chi2_stat:>10.2f}")

print(f"\n TEST RESULTS")
print(f"   Chi-Square (χ²) : {chi2_stat:.2f}")
print(f"   Degrees of Freedom (df) : {dof}")
print(f"   Critical Value (α=0.05) : {critical:.2f}")
print(f"   p-value          : {p_value:.6f}")

print(f"\n DECISION")
if chi2_stat > critical:
    print(f"   {chi2_stat:.2f} > {critical:.2f}")
    print("REJECT H₀ — Population HAS changed.")
    print("      There IS a significant population difference.")
    print("      Population has increased compared to 2010 census.")
else:
    print(f"   {chi2_stat:.2f} ≤ {critical:.2f}")
    print("ACCEPT H₀ — Population has NOT changed.")

print("=" * 55)
```
