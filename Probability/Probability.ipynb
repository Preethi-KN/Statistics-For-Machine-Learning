# Probability

```python
from fractions import Fraction
import math

# ── 1. BASIC PROBABILITY ──────────────────────────────────────
print("\n1. BASIC PROBABILITY ──────────────────────────────")
sample_space  = [1, 2, 3, 4, 5, 6]
even_outcomes = [x for x in sample_space if x % 2 == 0]
p_even = Fraction(len(even_outcomes), len(sample_space))

print(f"   Sample Space  : {sample_space}")
print(f"   Even outcomes : {even_outcomes}")
print(f"   P(even)       : {p_even}")
print(f"   Probability   : {p_even} = {float(p_even):.2f}")
```

## Additional Rule

```python
#Example — Dice rolled, what is P(even number OR number greater than 4)?
E1 = set([2, 4, 6])
E2 = set([5, 6])
intersection = E1 & E2
p_E1    = Fraction(len(E1), 6)
p_E2    = Fraction(len(E2), 6)
p_int   = Fraction(len(intersection), 6)
p_union = p_E1 + p_E2 - p_int
print(f"   P(even)        = {p_E1}")
print(f"   P(> 4)         = {p_E2}")
print(f"   P(even AND >4) = {p_int}")
print(f"   P(even OR >4)  = {p_E1} + {p_E2} - {p_int} = {p_union}")
```

## MULTIPLICATION RULE

```python
#Example — A deck of cards. What is the probability of drawing 2 Aces consecutively WITHOUT replacement?
p_ace1            = Fraction(4, 52) #52 cards--> 4 ace
p_ace2_given_ace1 = Fraction(3, 51) #51 cards-3 ace as 1 one already selected 
p_both_aces       = p_ace1 * p_ace2_given_ace1 # A ∩ B
print(f"   P(1st Ace)               = {p_ace1}")
print(f"   P(2nd Ace | 1st was Ace) = {p_ace2_given_ace1}")
print(f"   P(both Aces)             = {p_both_aces} ≈ {float(p_both_aces):.5f}")
```

## COMPLEMENT RULE 

```python
#Example — A coin is flipped twice. What is the probability of getting at least one head?
p_no_head           = Fraction(1, 2) * Fraction(1, 2)
p_at_least_one_head = 1 - p_no_head
print(f"   P(no heads in 2 flips)        = {p_no_head}")
print(f"   P(at least 1 head) = 1 - 1/4 = {p_at_least_one_head}")
```

## CONDITIONAL PROBABILITY 

```python
##A bag has 3 red balls and some blue balls. One ball is
# drawn and not replaced, another is drawn. What is the probability the
# second ball is red given the first ball is also red?
p_A         = Fraction(3, 5) # 3 red balls from total 5 balls
p_AnB       = Fraction(3, 5) * Fraction(2, 4) #1 ia already selected --> P(A∩B)
p_B_given_A = p_AnB / p_A
print(f"   P(1st red)               = {p_A}")
print(f"   P(1st AND 2nd red)       = {p_AnB}")
print(f"   P(2nd red | 1st was red) = {p_B_given_A}")
```
