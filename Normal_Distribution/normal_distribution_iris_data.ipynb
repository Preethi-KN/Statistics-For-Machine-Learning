```python
"""
Normal Distribution Analysis 
"""

import numpy as np
import seaborn as sns
# ── 1. DATASET ────────────────────────────────────────────────────────────────
df=sns.load_dataset("iris")
df.head
```

```python
# we are checking the normal distribution
sns.histplot(df['sepal_length'],kde=True) #kde=true => give you the curve
```

```python
# lets check for sepal width
sns.histplot(df['sepal_width'],kde=True)
```

```python
# Sepal length data is not in the normal distribution and sepal width looks like normal distribution
# so lets create the normal distridution using np.random.normal(mean,std,howmany data points)
new_sample=np.random.normal(0.5,0.2,1000)
new_sample
```

```python
#lets create the hisplot for new sample
sns.histplot(new_sample,kde=True)
```

```python
# we can al use log normal distribution, here we need to give mean and std
mu,sigma=3.,1.
log_Sample=np.random.lognormal(mu,sigma,100)
log_Sample
```

```python
sns.histplot(log_Sample,kde=True)
```

```python
sns.histplot(np.log(log_Sample),kde=True)
```

# Check the normal distribution

```python
#Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
    
```

```python
sample=np.random.normal(0.5,0.2,1000)
plot_data(sample) # if you get the stright line then its in normal distribution
```

```python
plot_data(df['sepal_width'])
```
