# Outliers ,IQR and Boxplot

```python
import numpy as np
import matplotlib.pyplot as plt

#data with outliers
dataset_val=[11,10,12,14,12,15,14,15,105,12,14,17,19,107,10,13,12,14,108,12,11,14,13,15,10,14,12,10,14,13,15,10]
```

```python
plt.hist(dataset_val)
```

# Z score

```python
#Z-score -> to find the find the outliers
outliers=[]
def find_outliers(data):
    threshold=3 #3 std
    mean=np.mean(data)
    std=np.std(data)

    for val in data:
        Z_score=(val-mean)/std
        if np.abs(Z_score)>threshold:
            outliers.append(val)
    return outliers
```

```python
find_outliers(dataset_val)
```

# IQR

```python
#IQR 
#sort the data
dataset_val=sorted(dataset_val)
dataset_val
```

```python
#Find q1 and q3 from the dataset
q1,q3=np.percentile(dataset_val,[25,75])
print(q1, q3)
```

```python
#Fine IQR=>q3-q2
IQR=q3-q1
print(IQR)
```

```python
# find the lower fance and upper fance
lower_fc=q1-(1.5*IQR) #formula=> q1-(1.5*IQR)
upper_fc=q3+(1.5*IQR) #formula=> q3+(1.5*IQR)

print(lower_fc,upper_fc)
```

```python
import seaborn as sns
sns.boxplot(dataset_val)
```
